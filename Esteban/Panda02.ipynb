```python
dfu['Time'] = pd.to_datetime(dfu['Time'])
```

```python
dfu.head(10)
```

```python
hinicio = time(6,0,0)
hfinal = time(12,0,0)
mask = (dfu['Time'].dt.time >= hinicio) & (dfu['Time'].dt.time <= hfinal)
```

```python
dfu6to12 = dfu.loc[mask]
```

```python
dfu6to12
```

```python
dfu6to12.count()
```

```python
def hu(hi, hf, mi=0, si=0, mf=0, sf=0):
    hinicio = time(hi, mi, si)
    hfinal = time(hf, mf, sf)
    mask = (dfu['Time'].dt.time >= hinicio) & (dfu['Time'].dt.time <= hfinal)
    return dfu.loc[mask]
```

```python
hu(23, 23, 0, 0, 59, 59)
```

```python
hu(6, 12)
```
