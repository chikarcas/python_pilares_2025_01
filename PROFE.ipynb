```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = np.arange(-10, 10, 0.01)
```

```python
y = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("Función seno")
plt.grid(True)
plt.show()
```
